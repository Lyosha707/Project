```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
EEG_DATA = pd.read_csv("EGG_DATA.csv")
```

```python
EEG_DATA_1 = EEG_DATA.copy()
EEG_DATA_1["Average Gamma (μV²)"] = EEG_DATA_1[["Gamma1 (μV²) - Lower 30-100 Hz of power spectrum", "Gamma2 (μV²) - Higher 30-100 Hz of power spectrum"]].mean(axis=1)
EEG_DATA_1["Average Alpha (μV²)"] = EEG_DATA_1[["Alpha1 (μV²) - Lower 8-11 Hz of power spectrum", "Alpha2 (μV²) - Higher 8-11 Hz of power spectrum"]].mean(axis=1)

EEG_DATA_1.head()
```

### Data Wrangling:

<span> Performed boolean masks on for loops to get desired video based on subject, or viceversa. Is Video[( subject #, video #)]</span>

```python
Video = {} #List to get each video based on subject

subjectIDs = EEG_DATA["SubjectID"].unique()
videoIDs = EEG_DATA["VideoID"].unique()

for i in subjectIDs:
    for j in videoIDs:
        Video[(i, j)] = EEG_DATA[(EEG_DATA["SubjectID"] == i) & (EEG_DATA["VideoID"] == j)].copy()
        Video[(i, j)]["Average Alpha (μV²)"] = Video[(i, j)][["Alpha1 (μV²) - Lower 8-11 Hz of power spectrum", "Alpha2 (μV²) - Higher 8-11 Hz of power spectrum"]].mean(axis=1)
        Video[(i, j)]["Average Beta (μV²) Amplitude 12-29 Hz of power spectrum)"] = Video[(i, j)][["Beta1 (μV²) - Lower 12-29 Hz of power spectrum", "Beta2 (μV²)  - Higher 12-29 Hz of power spectrum)"]].mean(axis=1)
        Video[(i, j)]["Average Gamma (μV²)"] = Video[(i, j)][["Gamma1 (μV²) - Lower 30-100 Hz of power spectrum", "Gamma2 (μV²) - Higher 30-100 Hz of power spectrum"]].mean(axis=1)
        Video[(i, j)]["Attention Dot size"] = (Video[(i, j)]["Attention - Measure of mental focus"])/1
        Video[(i, j)]["Calmness Dot size"] = (Video[(i, j)]["Mediation - Measure of calmness"])/1
```

```python
columns_to_average = [
    "Delta (μV²) - 1-3 Hz of power spectrum","Alpha2 (μV²) - Higher 8-11 Hz of power spectrum",
    "Alpha1 (μV²) - Lower 8-11 Hz of power spectrum",
    "Attention - Measure of mental focus",
    "Mediation - Measure of calmness"
]
EEG_DATA_video = EEG_DATA.groupby("VideoID")[columns_to_average].mean().reset_index().round(2)
EEG_DATA_video["VideoID"] = EEG_DATA_video["VideoID"].astype(int)
EEG_DATA_video["Alpha AVG (μV²) - 8-11 Hz of power spectrum"] = EEG_DATA_video[["Alpha1 (μV²) - Lower 8-11 Hz of power spectrum", "Alpha2 (μV²) - Higher 8-11 Hz of power spectrum"]].mean(axis=1)
```

```python
df = Video[(0, 0)].copy()

# Rename for convenience
df = df.rename(columns={
    "Delta (μV²) - 1-3 Hz of power spectrum": "Delta",
    "Average Alpha (μV²) Amplitude 11 Hz of power spectrum": "Alpha"
})

# Melt into long format
df_melted = df.melt(
    id_vars=["Time (Seconds)", "Calmness Dot size"],
    value_vars=[ "Alpha", "Beta", "Gamma"],
    var_name="Wave Type",
    value_name="Power (μV²)"
)

# Plot all waves on the same plot
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_melted,
    x="Time (Seconds)",
    y="Power (μV²)",
    hue="Wave Type",
    size="Calmness Dot size",
    sizes=(10, 200),
    alpha=0.5
)
plt.title("Brain Waves over Time for Subject 0, Video 0")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.show()
```
